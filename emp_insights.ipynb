```python
import pandas as pd
import numpy as np
df = pd.read_excel('myexcel.xlsx')
print(df.head())
print(df.info())
print(df.describe())
```

```python
df['height'] = np.random.randint(150, 181, df.shape[0])
```

```python
print(df.isnull().sum())
```

```python
team_distribution = df['Team'].value_counts()
team_percentage = (team_distribution / len(df)) * 100
print(team_distribution)
print(team_percentage)
```

```python
position_distribution = df['Position'].value_counts()
print(position_distribution)
```

```python
bins = [18, 25, 35, 45, 55, 65]
labels = ['18-24', '25-34', '35-44', '45-54', '55-64']
df['age_group'] = pd.cut(df['Age'], bins=bins, labels=labels)
age_group_distribution = df['age_group'].value_counts()
print(age_group_distribution)
```

```python
salary_by_team_position = df.groupby(['Team', 'Position'])['Salary'].sum()
highest_salary_expenditure = salary_by_team_position.idxmax(), salary_by_team_position.max()
print(highest_salary_expenditure)
```

```python
correlation = df['Age'].corr(df['Salary'])
print(f"Correlation between age and salary: {correlation}")
```

```python
import matplotlib.pyplot as plt

team_distribution.plot(kind='bar')
plt.title('Distribution of Employees Across Teams')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.show()
```

```python
position_distribution.plot(kind='pie', autopct='%1.1f%%')
plt.title('Employee Distribution by Position')
plt.ylabel('')
plt.show()
```

```python
age_group_distribution.plot(kind='bar')
plt.title('Predominant Age Group Among Employees')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

```python
age_group_distribution.plot(kind='bar')
plt.title('Predominant Age Group Among Employees')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

```python
salary_by_team_position.plot(kind='bar', figsize=(20, 10)) 
plt.title('Salary Expenditure by Team and Position')
plt.xlabel('Team, Position')
plt.ylabel('Total Salary Expenditure')
plt.xticks(rotation=90, ha='right')  
plt.tight_layout()  
plt.show()

```

```python
plt.scatter(df['Age'], df['Salary'])
plt.title('Correlation Between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
